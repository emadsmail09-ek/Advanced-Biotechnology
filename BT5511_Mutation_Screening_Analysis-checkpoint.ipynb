```python
import glob

files = glob.glob("*.txt")
print(files)
```

```python
import pandas as pd

data_list = []

for file in files:
    df = pd.read_csv(file, sep="\t")
    data_list.append(df)

combined_df = pd.concat(data_list, ignore_index=True)

print(combined_df.shape)
```

```python
combined_df["mRNA_WT_mean"] = combined_df[
    ["mRNA.Expression.WT.Rep1",
     "mRNA.Expression.WT.Rep2",
     "mRNA.Expression.WT.Rep3"]
].mean(axis=1)

combined_df["mRNA_Mut_mean"] = combined_df[
    ["mRNA.Expression.Mut.Rep1",
     "mRNA.Expression.Mut.Rep2",
     "mRNA.Expression.Mut.Rep3"]
].mean(axis=1)

combined_df[["mRNA_WT_mean", "mRNA_Mut_mean"]].head()
```

```python
combined_df["Protein_WT_mean"] = combined_df[
    ["Protein.Expression.WT.Rep1",
     "Protein.Expression.WT.Rep2",
     "Protein.Expression.WT.Rep3"]
].mean(axis=1)

combined_df["Protein_Mut_mean"] = combined_df[
    ["Protein.Expression.Mut.Rep1",
     "Protein.Expression.Mut.Rep2",
     "Protein.Expression.Mut.Rep3"]
].mean(axis=1)

combined_df[["Protein_WT_mean", "Protein_Mut_mean"]].head()
```

```python
combined_df["Viability_WT_mean"] = combined_df[
    ["CellViability.WT.Rep1",
     "CellViability.WT.Rep2",
     "CellViability.WT.Rep3"]
].mean(axis=1)

combined_df["Viability_Mut_mean"] = combined_df[
    ["CellViability.Mut.Rep1",
     "CellViability.Mut.Rep2",
     "CellViability.Mut.Rep3"]
].mean(axis=1)

combined_df[["Viability_WT_mean", "Viability_Mut_mean"]].head()
```

```python
combined_df["mRNA_change"] = combined_df["mRNA_Mut_mean"] - combined_df["mRNA_WT_mean"]

combined_df["Protein_change"] = combined_df["Protein_Mut_mean"] - combined_df["Protein_WT_mean"]

combined_df["Viability_change"] = combined_df["Viability_Mut_mean"] - combined_df["Viability_WT_mean"]

combined_df[["mRNA_change", "Protein_change", "Viability_change"]].head()
```

```python
def detect_mutation(wt, mut):
    if len(wt) == len(mut):
        return "Substitution"
    elif len(wt) < len(mut):
        return "Insertion"
    elif len(wt) > len(mut):
        return "Deletion"
    else:
        return "Unknown"

combined_df["Mutation_Type"] = combined_df.apply(
    lambda row: detect_mutation(row["WildType.Sequence"], row["Mutant.Sequence"]),
    axis=1
)

combined_df[["Gene", "Mutation_Type"]].head()
```

```python
top5 = combined_df.sort_values(by="Viability_change").head(5)

top5[["Gene", "mRNA_change", "Protein_change", "Viability_change"]]
```

```python
# Create figure
fig, axes = plt.subplots(2, 2, figsize=(12,10))

# Colors for mutation types
colors = {
    "Substitution": "orange",
    "Deletion": "blue",
    "Insertion": "green"
}

# --- 1. BOXPLOT (using mean values for better comparison) ---
data = []
labels = []

for mtype in combined_df["Mutation_Type"].unique():
    subset = combined_df[combined_df["Mutation_Type"] == mtype]
    data.append(subset["mRNA_WT_mean"])
    labels.append(mtype)

axes[0,0].boxplot(data, tick_labels=labels)
axes[0,0].set_title("mRNA Expression by Mutation Type")
axes[0,0].set_yscale('log')  # KEY improvement

# --- 2. SCATTER (log scale like example) ---
for mtype in combined_df["Mutation_Type"].unique():
    subset = combined_df[combined_df["Mutation_Type"] == mtype]
    axes[0,1].scatter(
        subset["mRNA_Mut_mean"],
        subset["Protein_Mut_mean"],
        label=mtype,
        color=colors[mtype]
    )

axes[0,1].set_title("mRNA vs Protein (Mutant)")
axes[0,1].set_xlabel("mRNA Mutant Mean")
axes[0,1].set_ylabel("Protein Mutant Mean")
axes[0,1].set_xscale('log')   # IMPORTANT
axes[0,1].set_yscale('log')   # IMPORTANT
axes[0,1].legend()

# --- 3. KDE (smooth density instead of histogram) ---
for mtype in combined_df["Mutation_Type"].unique():
    subset = combined_df[combined_df["Mutation_Type"] == mtype]
    subset["Viability_Mut_mean"].plot(
        kind='kde',
        ax=axes[1,0],
        label=mtype,
        color=colors[mtype]
    )

axes[1,0].set_title("Viability Distribution (Density)")
axes[1,0].set_xlabel("Viability Mutant Mean")
axes[1,0].legend()

# --- Empty subplot ---
axes[1,1].axis("off")

plt.tight_layout()
plt.show()
```
